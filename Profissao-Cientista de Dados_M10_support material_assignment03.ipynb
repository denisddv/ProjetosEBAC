1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('previsao_de_renda.csv')

df['data_ref'] = pd.to_datetime(df['data_ref'])


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.countplot(x=df['data_ref'].dt.year, hue='posse_de_imovel', data=df)
plt.title('Distribuição de Posse de Imóvel no Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Posse de Imóvel')


plt.subplot(1, 2, 2)
sns.countplot(x=df['data_ref'].dt.year, hue='posse_de_veiculo', data=df)
plt.title('Distribuição de Posse de Veículo no Tempo')
plt.xlabel('Ano')
plt.ylabel('Contagem')
plt.legend(title='Posse de Veículo')

plt.tight_layout()
plt.show()
````

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
df['ano'] = df['data_ref'].dt.year


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.pointplot(x='ano', y='renda', hue='posse_de_imovel', data=df, errorbar='sd', dodge=True)
plt.title('Média de Renda por Posse de Imóvel no Tempo')
plt.xlabel('Ano')
plt.ylabel('Média de Renda')
plt.legend(title='Posse de Imóvel')


plt.subplot(1, 2, 2)
sns.pointplot(x='ano', y='renda', hue='posse_de_veiculo', data=df, errorbar='sd', dodge=True)
plt.title('Média de Renda por Posse de Veículo no Tempo')
plt.xlabel('Ano')
plt.ylabel('Média de Renda')
plt.legend(title='Posse de Veículo')

plt.tight_layout()
plt.show()
````
