####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sinasc_df = pd.read_csv('sinasc_RO_2019.csv')


plt.figure(figsize=(8, 6))
sns.boxplot(x='IDADEMAE', data=sinasc_df, color='skyblue')
plt.title('Boxplot da Idade da Mãe')
plt.xlabel('Idade da Mãe')


plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='PESO', data=sinasc_df, color='skyblue')
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso do Bebê (gramas)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=sinasc_df, x='IDADEMAE', bins=20, kde=True, color='skyblue')
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê
