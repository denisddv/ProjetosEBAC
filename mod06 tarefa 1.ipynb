#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sinasc_df = pd.read_csv('SINASC_RO_2019.csv')


sinasc_df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(12, 8))


plt.subplot(2, 3, 1)
sinasc_df['APGAR1'].value_counts().plot(kind='bar', color='skyblue')
plt.title('APGAR1')
plt.xlabel('Pontuação')
plt.ylabel('Número de registros')


plt.subplot(2, 3, 2)
sinasc_df['APGAR5'].value_counts().plot(kind='bar', color='salmon')
plt.title('APGAR5')
plt.xlabel('Pontuação')
plt.ylabel('Número de registros')


plt.subplot(2, 3, 3)
sinasc_df['SEXO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('SEXO')
plt.xlabel('Sexo')
plt.ylabel('Número de registros')


plt.subplot(2, 3, 4)
sinasc_df['GESTACAO'].value_counts().plot(kind='bar', color='purple')
plt.title('GESTACAO')
plt.xlabel('Tempo de gestação')
plt.ylabel('Número de registros')


plt.subplot(2, 3, 5)
sinasc_df['ESTCIVMAE'].value_counts().plot(kind='bar', color='orange')
plt.title('ESTCIVMAE')
plt.xlabel('Estado civil da mãe')
plt.ylabel('Número de registros')


plt.tight_layout()


plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
avg_apgar5_by_consultas = sinasc_df.groupby('CONSULTAS')['APGAR5'].mean().reset_index()


avg_apgar5_by_consultas = avg_apgar5_by_consultas.sort_values(by='CONSULTAS')


plt.figure(figsize=(10, 6))
plt.bar(avg_apgar5_by_consultas['CONSULTAS'], avg_apgar5_by_consultas['APGAR5'], color='skyblue')
plt.title('APGAR5 Médio por Número de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.xticks(avg_apgar5_by_consultas['CONSULTAS'])
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.subplot(2, 2, 1)
sns.countplot(x='SEXO', data=sinasc_df, palette='pastel')
plt.title('Frequência de SEXO')


plt.subplot(2, 2, 2)
sns.countplot(x='GESTACAO', data=sinasc_df, palette='pastel')
plt.title('Frequência de GESTACAO')


plt.subplot(2, 2, 3)
sns.countplot(x='ESTCIVMAE', data=sinasc_df, palette='pastel')
plt.title('Frequência de ESTCIVMAE')


plt.subplot(2, 2, 4)
sns.countplot(x='CONSULTAS', data=sinasc_df, palette='pastel')
plt.title('Frequência de CONSULTAS')

plt.tight_layout()

plt.show()
```
