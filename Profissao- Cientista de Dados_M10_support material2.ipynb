```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
df = pd.read_csv('previsao_de_renda.csv')

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.pointplot(x='posse_de_imovel', y='renda', data=df, capsize=0.2, errorbar='sd')
plt.title('Point Plot de Posse de Imóvel vs Renda')
plt.xlabel('Posse de Imóvel')
plt.ylabel('Renda')


plt.subplot(1, 2, 2)
sns.pointplot(x='posse_de_veiculo', y='renda', data=df, capsize=0.2, errorbar='sd')
plt.title('Point Plot de Posse de Veículo vs Renda')
plt.xlabel('Posse de Veículo')
plt.ylabel('Renda')

plt.tight_layout()
plt.show()
```
